### Import Libraries

````python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

### Dataset Loading

````python
# Load Dataset
credit=pd.read_csv('credits.csv')
credit.head()
title=pd.read_csv('titles.csv')
title.head()
````

### Merging both data

````python
# Both data are merged
df=pd.merge(credit,title,on='id')
````

````python
# Dataset First Look
df.head()
````

### Dataset Rows & Columns count

````python
# Dataset Rows & Columns count
df.shape
````

````python
# Missing Values/Null Values Count
df.isnull().sum()                                 # isnull() use to check for null values
                                                  # sum() use of count total values
````
